```python
## check for null values
df.isna().sum()
```

```python
## check for data type of the column and fill null values with
## mean for numeric data and mode for object data

for col in df.columns:
    if df[col].dtype == object:
      # Fill with mode for object columns (strings)
      df[col].fillna(df[col].mode().iloc[0], inplace=True)
    elif pd.api.types.is_numeric_dtype(df[col]):
      # Fill with mean for integer columns
      df[col].fillna(df[col].mean(), inplace=True)
```

```python
## check if above works
df.isna().sum()
```

```python
# Define table name
table_name = "car_table"

# Create table if it doesn't exist (adjust data types as needed)
df.to_sql(table_name, engine, index=False, if_exists='append')

print(f"Data successfully stored in table: {table_name}")
```
